```python
import requests
import pandas as pd
from google.colab import data_table
from IPython.display import clear_output
import time

def show_advanced_crypto_dashboard():
    # Research Task 1 & 2: Currency INR rakhi hai aur limit 20 coins kar di hai
    url = "https://api.coingecko.com/api/v3/coins/markets"
    params = {
        'vs_currency': 'inr',
        'order': 'market_cap_desc',
        'per_page': 20,      # Top 20 coins
        'page': 1,
        'sparkline': False
    }

    try:
        while True:
            response = requests.get(url, params=params)
            data = response.json()

            # Data Processing
            df = pd.DataFrame(data)[['name', 'symbol', 'current_price', 'price_change_percentage_24h', 'market_cap']]

            # Research Task 3: USD Conversion (Approx calculation for comparison)
            # Maan lete hain 1 USD = 83 INR
            df['Price (INR)'] = df['current_price'].apply(lambda x: f"₹{x:,.2f}")
            df['Price (USD)'] = df['current_price'].apply(lambda x: f"${(x/83):,.2f}")

            # Logic for Alert: Agar change negative hai toh hum table mein dekh sakte hain
            df['Status'] = df['price_change_percentage_24h'].apply(lambda x: "🔻 DOWN" if x < 0 else "🟢 UP")
            df['24h Change (%)'] = df['price_change_percentage_24h'].apply(lambda x: f"{x:.2f}%")

            # Final Table Columns select karna
            final_df = df[['name', 'Price (INR)', 'Price (USD)', '24h Change (%)', 'Status']]
            final_df.columns = ['Coin Name', 'Price (Indian Rupee)', 'Price (US Dollar)', '24h Change', 'Market Trend']

            # Display Dashboard
            clear_output(wait=True)
            print(f"🇮🇳 PRAGATI'S LOCALIZED CRYPTO TRACKER 🚀")
            print(f"Status: Live | Last Update: {time.strftime('%H:%M:%S')}")
            print("-" * 50)

            # Colab's interactive table
            display(data_table.DataTable(final_df, include_index=False, num_rows_per_page=10))

            # 1 minute ka wait taaki API block na ho (Research: API limits)
            time.sleep(60)

    except Exception as e:
        print(f"Error: {e}")
        print("API limit reach ho gayi hai, 1 minute wait karein...")
        time.sleep(60)

# Dashboard ko chalu karein
show_advanced_crypto_dashboard()
```
